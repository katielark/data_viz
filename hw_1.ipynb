Modify the function “attributes of service” from your in-class work into a new function that take the service usage data frame as input, and return a dataframe, with rows for services, and additional columns for total number of recipients and average monthly number of recipients.
Note: We used chatgpt on this code. We originally calculated the avg_monthly inside the agg() command. Chatgpt suggested we move the column to outside the command and add the df[]. This makes sense because the columns total_recipients and distinct_month had not been defined. 
Notably, the avg_monthly uses the total_recipients and not the total_unique_recipients column (which have very different values)

```python
def get_services_attribute(df):
    services = df.groupby(df['service'].cat.remove_unused_categories()).agg(
        total_recipients = ('service', 'count'),
        total_unique_recpients = ('id', 'nunique'),
        distinct_month = ('month', 'nunique'),
    ).reset_index()
    
    services['avg_monthly'] = services['total_recipients'] / services['distinct_month']
    
    return services
```

Write another function that produces the numbers of recipients by month.

```python
def get_recipient_month_attribute(df):
    service_month = df.groupby(['service', 'month']).agg(
        num_id = ('id', 'count'),
    ).reset_index()

    return service_month
```
